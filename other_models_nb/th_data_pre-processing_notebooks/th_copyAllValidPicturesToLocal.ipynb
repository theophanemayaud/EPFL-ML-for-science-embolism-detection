# Import all pictures that are used to train into local folder

From the nicely prepared file analysis_overview_df_with_pathnames.csv, we'll use the pathname to copy to local subfolder

```python
import pandas as pd

import shutil
```

```python
analysis_overview_df = pd.read_csv("./th_sampledata/analysis_overview_df_with_pathnames.csv")
```

```python
analysis_overview_df
```

```python
# We'll copy with the structure destination names 
# SessionX_[living/flushed]_[plan name]_[image name] 
# to find it more easily later
for index, row in analysis_overview_df.iterrows():
    srcpath = row.pathname
    dstpath = f"./th_analysedimages/{row.session}_{row.scan_nb}_{row.plant_name}_{row.image_nb}.tif"
    print(srcpath + "\n"+ dstpath+"\n\n")
    shutil.copy(srcpath, dstpath, follow_symlinks=True)
```
